```python
import pandas as pd 
```

```python
df = pd.read_csv("../datasets/dataset.csv")
```

```python
df
```

```python
complete_map_data = {}

for index,row in df.iterrows():
    dis = row["Disease"]
    if dis not in complete_map_data:
        complete_map_data[dis] = set()
    all_symptons_col = df.columns[1:]
    for col in all_symptons_col:
        symp = row[col]
        if pd.notna(symp):
            symp = symp.strip().lower().replace("_"," ")
            complete_map_data[dis].add(symp)

all_text = ""

for key in complete_map_data:
    symps = complete_map_data[key]
    symp_list = list(symps)
    symp_text = ",".join(symp_list)
    all_text+=f"Disease Name: {key} and symptoms are {symp_text}\n"


with open("disease_text.txt","w") as file:
    file.write(all_text)
```
